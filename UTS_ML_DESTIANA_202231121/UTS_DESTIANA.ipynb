```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt ##menampilkan library
```

```python
data1 = pd.read_csv('K01_heart_disease.csv')
data1 ##menampilkan data set
```

```python
data1.info() ##menampilkan informasi tentang objek data1
```

```python
categorical_features = [feature for feature in data1.columns
                       if data1[feature].dtypes == "O"]
data1[categorical_features] ## Memilih Fitur Kategorikal dari DataFrame
```

```python
for i in categorical_features:
    print(data1[i].unique()) ##menampilkan nilai unik dari fitur-fitur kategorikal dalam sebuah dataset
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for feature in categorical_features:
    print(data1[feature].unique())
    data1[feature] = label_encoder.fit_transform(data1[feature])
    print(data1[feature].unique()) ##untuk mengonversi data kategorikal dalam sebuah dataset menjadi data numerik
```

```python
data1 ##menampilkan data
```

```python
x = data1.drop(['Age', 'HeartDisease'], axis = 1) #berisi semua fitur input untuk model (variabel independen).
y = data1['Age'] #y berisi data target (output) yang ingin diprediksi (variabel dependen).
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0) ##data untuk membantu memvalidasi performa model dan mengurangi risiko overfitting.
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train) ##digunakan untuk membuat model klasifikasi Naive Bayes berbasis distribusi Gaussian.
```

```python
y_pred_train = model.predict(x_train) ##digunakan untuk mengukur seberapa baik model memprediksi data pelatihan. Hal ini bisa membantu mendeteksi masalah overfitting jika model sangat akurat di data pelatihan tetapi tidak di data pengujian.
y_pred_test = model.predict(x_test) ##digunakan untuk mengevaluasi performa model pada data pengujian, yang memberikan gambaran tentang seberapa baik model akan bekerja pada data baru.
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print('evaluasi training')
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))

print('evaluasi testing')
print(confusion_matrix(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test)) ##laporan klasifikasi untuk data pelatihan menunjukkan metrik yang tinggi, 
##tetapi laporan untuk data pengujian menunjukkan metrik yang lebih rendah, ini dapat menunjukkan masalah overfitting.
```
